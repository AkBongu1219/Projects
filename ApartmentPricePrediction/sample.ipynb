```python
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import math as m
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn import svm, datasets
from sklearn.datasets import load_digits
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso, Ridge
import matplotlib
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
```

```python
df = pd.read_csv('C:\\Users\\Guest_User\\Downloads\\vehicle_data\\vehicle_maintenance_data.csv')
df.head()
```

```python
df['Last_Service_Date'] = pd.to_datetime(df['Last_Service_Date'])
df['Warranty_Expiry_Date'] = pd.to_datetime(df['Warranty_Expiry_Date'])
df.head()
```

```python
df['Last_Service_Year'] = df['Last_Service_Date'].dt.year
df['Last_Service_Month'] = df['Last_Service_Date'].dt.month
df['Last_Service_Day'] = df['Last_Service_Date'].dt.day
df['Last_Service_DayOfWeek'] = df['Last_Service_Date'].dt.dayofweek
df.head()
```

```python
df['Warranty_Expiry_Year'] = df['Warranty_Expiry_Date'].dt.year
df['Warranty_Expiry_Month'] = df['Warranty_Expiry_Date'].dt.month
df['Warranty_Expiry_Day'] = df['Warranty_Expiry_Date'].dt.day
df['Warranty_Expiry_DayOfWeek'] = df['Warranty_Expiry_Date'].dt.dayofweek
df.head()
```

```python
df.drop(['Last_Service_Date', 'Warranty_Expiry_Date'], axis=1, inplace=True)
df.head()
```

```python
X = df.drop(['Fuel_Type','Owner_Type','Insurance_Premium','Fuel_Efficiency'], axis = 'columns')
dummies = pd.get_dummies(X)
X = dummies
X.head()
```

```python
len(X)
```

```python
y = df['Need_Maintenance']
y.head()
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.25, random_state=10)
xtrain.head()
```

```python
xtrain['Last_Service_Month_Sin'] = np.sin(2 * np.pi * xtrain['Last_Service_Month'] / 12)
xtrain['Last_Service_Month_Cos'] = np.cos(2 * np.pi * xtrain['Last_Service_Month'] / 12)
xtest['Last_Service_Month_Sin'] = np.sin(2 * np.pi * xtest['Last_Service_Month'] / 12)
xtest['Last_Service_Month_Cos'] = np.cos(2 * np.pi * xtest['Last_Service_Month'] / 12)
xtrain.head()
```

```python
model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
score = model.score(xtest, ytest)
print(f"Model score on test set: {score}")
```

```python
cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
scores = cross_val_score(model, X, y, cv=cv)
```

```python
print(f"Cross-validation scores: {scores}")
print(f"Mean cross-validation score: {np.mean(scores)}")
```

```python
algos = {
    'linear_regression': {
        'model': LinearRegression(),
        'params': {
            #'normalize': [True, False]
        }
    },
    'lasso': {
        'model': Lasso(),
        'params': {
            'alpha': [1, 2],
            'selection': ['random', 'cyclic']
        }
    },
    'decision_tree': {
        'model': DecisionTreeRegressor(),
        'params': {
            'criterion': ['absolute_error', 'poisson', 'friedman_mse', 'squared_error'],
            'splitter': ['best', 'random']
        }
    }
}
```

```python
scores = []
cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
```

```python
for algo_name, config in algos.items():
    gs = GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
    gs.fit(X, y)
    scores.append({
        'model': algo_name,
        'best_score': gs.best_score_,
        'best_params': gs.best_params_
    })
```
